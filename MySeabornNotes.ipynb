```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.barplot("sex", "tip", data=tips, estimator=np.sum);
```

```python
sns.barplot("sex", "tip", data=tips, estimator=np.min);
```

```python
sns.barplot("sex", "tip", data=tips, estimator=np.mean, ci="sd");
```

```python
sns.barplot("sex", "tip", data=tips, estimator=np.mean, ci="sd", hue="smoker");
```

```python
fig, ax_tip= plt.subplots()
ax_tip= sns.countplot(x="day", data=tips);

for p in ax_tip.patches:
    ax_tip.annotate("("+
                   str(p.get_height()) +
                   ")" +
                   "%" +
                   str(round((p.get_height()) / tips.day.count(), 2)*100), (p.get_x()+0.07, p.get_height()-10));
```

```python
fig, ax= plt.subplots(nrows=1, ncols=2, figsize=[16,6])

sns.barplot(data=tips, x="day", y="total_bill", ax=ax[0])
sns.countplot(data=tips, x="day", ax=ax[1]);
```

```python
fig,ax= plt.subplots(nrows=1, ncols=2, figsize=(16,6))

sns.barplot(data=tips, x="day", y="total_bill", ax=ax[0], estimator=np.median)
sns.countplot(data=tips, x="day", ax=ax[1]);
```
